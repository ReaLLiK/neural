```python
from keras.applications.vgg16 import VGG16
from keras.applications.resnet50 import ResNet50, preprocess_input, decode_predictions
from keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input, decode_predictions
import matplotlib.pyplot as plt
import numpy as np
```

```python
model = VGG16(weights='imagenet')
```

```python
img = image.load_img('mycar.jpg', target_size=(224, 224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
```

```python
preds = model.predict(x)
print('Результаты распознавания:', decode_predictions(preds, top=5)[0])
plt.imshow(img)
```

```python
model2 = ResNet50(weights='imagenet')
```

```python
preds2 = model2.predict(x)
print('Результаты распознавания:', decode_predictions(preds2, top=5)[0])
plt.imshow(img)
```
